<h3>Задание 1</h3>

```python
from datetime import datetime as dt

moscow_times = 'Wednesday, October 2, 2002'
dt_moscow_times = dt.strptime(moscow_times, '%A, %B %d, %Y')
print(dt_moscow_times)

guardian = 'Friday, 11.10.13'
dt_guardian = dt.strptime(guardian, '%A, %d.%m.%y')
print(dt_guardian)

daily_news = 'Thursday, 18 August 1977'
dt_daily_news = dt.strptime(daily_news, '%A, %d %B %Y')
print(dt_daily_news)
```

<h3>Задание 3</h3>

```python
from datetime import datetime as dt
from datetime import timedelta

start_date = input('Введите дату начала в формате YYYY-MM-DD: ')
end_date = input('Введите дату конца в формате YYYY-MM-DD: ')

period = []

try:
    dt_start_date = dt.strptime(start_date, '%Y-%m-%d')
    dt_end_date = dt.strptime(end_date, '%Y-%m-%d')
    
    current_date = dt_start_date
    
    while current_date <= dt_end_date:
        period.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)
    
    print(period)
except:
    print(period)
```

1. Что значит ошибка list index out of range?
В коде происходит обращение по индексу к элементу списка, которого нет в списка - индекс списка вне диапазона.

2. Почему при первом запуске функция работает корректно, а при втором — нет?
Функция возвращает второй с конца элемент списка. При этом на вход функции необязательно подавать список, в котором необходимо удалять элементы. Значение списка задается по умолчанию, если не передается на вход функции. При первом запуске функции из списка из 3 элементов удаляется 1 элемент, возвращается 2-ой элемент. Но список - изменяемый тип данных, поэтому список, который идет на вход функции по умолчанию, так же меняется. Так что при втором запуске функции на вход уже поступает список из 2 элементов, удаляется 1 элемент. Остается всего 1 элемент в списке. И функция пытается вернуть 2-ой элемент в списке (который отсутствует). Возникает ошибка, что индекс списка вне диапазона.
