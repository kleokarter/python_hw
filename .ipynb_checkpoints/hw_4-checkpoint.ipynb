```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def p(number = None):
    if number == None:
        number = input('Введите номер документа: ')
    else:
        number = str(number)
    doc_owner = ''
    result = 'Документ не найден в базе'
    for doc in documents:
        if doc['number'] == number:
            doc_owner = doc['name']
    if doc_owner != '':
        result = f'Владелец документа: {doc_owner}'
        
    return result
```

```python
def s(number = None):
    if number == None:
        number = input('Введите номер документа: ')
        doc_shelf = ''
        result = 'Документ не найден в базе'
        for shelf in directories:
            if number in directories[shelf]:
                doc_shelf = shelf
        if doc_shelf != '':
            result = f'Документ хранится на полке: {doc_shelf}'
    else:
        number = str(number)
        doc_shelf = ''
        result = 'Документ не найден в базе'
        for shelf in directories:
            if number in directories[shelf]:
                doc_shelf = shelf
        if doc_shelf != '':
            result = doc_shelf
    
    return result
```

```python
def l():
    result = ''
    for doc in documents:
        result += f"№ {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {s(doc['number'])} \n" 
    return result
```

```python
def ads(new_shelf = None):
    if new_shelf == None:
        new_shelf = input('Введите номер полки: ')
    else:
        new_shelf = str(new_shelf)
    result = f"Такая полка уже существует. Текущий перечень полок: {', '.join(list(directories.keys()))}"
    if new_shelf not in directories.keys():
        directories[new_shelf] = list()
        result = f"Полка добавлена. Текущий перечень полок: {', '.join(list(directories.keys()))}"
    
    return result
```

```python
def ds(delete_shelf = None):
    if delete_shelf == None:
        delete_shelf = input('Введите номер полки: ')
    else:
        delete_shelf = str(new_shelf)
    result = f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(list(directories.keys()))}"
    if delete_shelf in directories.keys() and len(directories[delete_shelf]) == 0:
        directories.pop(delete_shelf)
        result = f"Полка удалена. Текущий перечень полок: {', '.join(list(directories.keys()))}"
    
    return result
```

```python
def ad():
    new_doc_number = input('Введите номер документа: ')
    new_doc_type = input('Введите тип документа: ')
    new_doc_owner = input('Введите владельца документа: ')
    new_doc_shelf = input('Введите полку для хранения: ')
    if new_doc_shelf in directories.keys():
        directories[new_doc_shelf] = new_doc_number
        new_document = {'type': new_doc_type, 'number': new_doc_number, 'name': new_doc_owner}
        documents.append(new_document)
        return f"Документ добавлен. Текущий список документов:\n{get_func('l')}"
    else:
        return f"Такой полки не существует. Добавьте полку командой as. Текущий список документов:\n{get_func('l')()}"
```

```python
def d(number = None):
    if number == None:
        number = input('Введите номер документа: ')
    else:
        number = str(number)
    
    result = f"Документ не найден в базе.\nТекущий список документов:\n{get_func('l')()}"
    delete_doc_number = ''
    
    for doc in documents:
        if doc['number'] == number:
            delete_doc_number = number
            
        if delete_doc_number != '':
            documents.remove(doc)
            directories.pop(s(number))
            result = f"Документ удален.\nТекущий список документов:\n{get_func('l')()}"
    
    return result
```

```python
def m():
    number = input('Введите номер документа: ')
    new_shelf = input('Введите номер полки: ')
    
    move_doc_number = ''
    result = f"Такой полки не существует. Текущий перечень полок: {', '.join(list(directories.keys()))}"
    
    if new_shelf not in directories.keys():
        return result
    else:
        for doc in documents:
            if doc['number'] == number:
                move_doc_number = number
            
            if move_doc_number != '':
                directories[s(number)].remove(move_doc_number)
                directories[new_shelf].append(move_doc_number)
                result = f"Документ перемещен.\nТекущий список документов:\n{get_func('l')()}"
            else:
                result = f"Документ не найден в базе.\nТекущий список документов:\n{get_func('l')()}"
    
    return result
```

```python
def get_func(command):
    func_dict = {
        'p': p,
        's': s,
        'l': l,
        'ads': ads,
        'ds': ds,
        'ad': ad,
        'd': d,
        'm': m
    }
    
    return(func_dict[command])
```

```python
def main():
    command = ''
    while command != 'q':
        command = input('Введите команду: ')
        if command != 'q':
            print(get_func(command)())
```

```python
main()
```
