<h3>Задание 1</h3>

```python
import json

purchases = {}

f = open('purchase_log.txt')

content = f.readlines()

for i, line in enumerate(content):
    line = line.strip()
    
    if i > 0:
            purchases[json.loads(line)['user_id']] = json.loads(line)['category']  
```

```python
f.close()
```

<h3>Задание 2</h3>

```python
purchases_categ = []

with open('visit_log.csv') as f:
    with open('funnel.csv', 'w') as funnel:
        for i, line in enumerate(f):
        
            if i == 0:
                line = line.strip().split(',')
                line.append('category')
                new_line = ','.join(line)
                funnel.write(new_line+'\n')
            else:
                line = line.strip().split(',')
                category = purchases.get(line[0], 'no category')
                
            
                if category != 'no category':
                    line.append(category)
                    new_line = ','.join(line)
                    funnel.write(new_line + '\n')
```
