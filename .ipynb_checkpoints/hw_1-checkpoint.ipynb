```python
phrase1 = input('Введите первую фразу: ')
phrase2 = input('Введите вторую фразу: ')

if len(phrase1) > len(phrase2):
    print('Фраза 1 длиннее фразы 2')
elif len(phrase1) < len(phrase2):
    print('Фраза 2 длиннее фразы 1')
else:
    print('Фразы равной длины')
```

<h3>Задание 3</h3>

```python
birth_day = int(input('Введите день: '))
birth_month = (input('Введите месяц: ')).upper()

if birth_month == 'ЯНВАРЬ':
    if birth_day < 20:
        zodiac_sign = 'Козерог' 
    else: 
        zodiac_sign = 'Водолей'
elif birth_month == 'ФЕВРАЛЬ':
    if birth_day < 19:
        zodiac_sign = 'Водолей'
    else:
        zodiac_sign = 'Рыбы'
elif birth_month == 'МАРТ':
    if birth_day < 21:
        zodiac_sign = 'Рыбы' 
    else:
        zodiac_sign = 'Овен'
elif birth_month == 'АПРЕЛЬ':
    if birth_day < 21:
        zodiac_sign = 'Овен' 
    else: zodiac_sign = 'Телец'
elif birth_month == 'МАЙ':
    if birth_day < 22:
        zodiac_sign = 'Телец' 
    else:
        zodiac_sign = 'Близнецы'
elif birth_month == 'ИЮНЬ':
    if birth_day < 22:
        zodiac_sign = 'Близнецы'
    else:
        zodiac_sign = 'Рак'
elif birth_month == 'ИЮЛЬ':
    if birth_day < 24:
        zodiac_sign = 'Рак'
    else:
        zodiac_sign = 'Лев'
elif birth_month == 'АВГУСТ':
    if birth_day < 24:
        zodiac_sign = 'Лев'
    else:
        zodiac_sign = 'Дева'
elif birth_month == 'СЕНТЯБРЬ':
    if birth_day < 24:
        zodiac_sign = 'Дева'
    else:
        zodiac_sign = 'Весы'
elif birth_month == 'ОКТЯБРЬ':
    if birth_day < 25:
        zodiac_sign = 'Весы'
    else:
        zodiac_sign = 'Скорпион'
elif birth_month == 'НОЯБРЬ':
    if birth_day < 23:
        zodiac_sign = 'Скорпион'
    else:
        zodiac_sign = 'Стрелец'
elif birth_month == 'ДЕКАБРЬ':
    if birth_day < 23:
        zodiac_sign = 'Стрелец'
    else:
        zodiac_sign = 'Козерог'
        
print(zodiac_sign)
```

<h3>Задание 4</h3>

```python
width = int(input('Введите ширину: '))
length = int(input('Введите длину: '))
height = int(input('Введите высоту: '))

if width <= 15 and length <= 15 and height <= 15:
    print('Коробка №1')
elif width >= 200 or length >= 200 or height >= 200:
    print('Упаковка для лыж')
elif (width > 15 and width <= 50) or (length  > 15 and length <= 50) or (height > 15 and height >= 50):
    print('Коробка №2')
else:
    print('Коробка №3')
```

<h3>Задание 5</h3>

```python
ticket_number = input('Введите номер билета: ')

if (int(ticket_number[0]) + int(ticket_number[1]) + int(ticket_number[2])) == (int(ticket_number[3]) + int(ticket_number[4]) + int(ticket_number[5])):
    print('Счастливый билет')
else:
    print('Несчастливый билет')
```

<h3>Задание 6</h3>

```python
import math

figure = (input('Введите тип фигуры: ')).upper()

if figure == 'КРУГ':
    radius = int(input('Введите радиус: '))
    print('Площадь круга:', math.pi * (radius ** 2))
elif figure == 'ТРЕУГОЛЬНИК':
    side1 = int(input('Введите длину первой стороны: '))
    side2 = int(input('Введите длину второй стороны: '))
    side3 = int(input('Введите длину третьей стороны: '))
    half_per = (side1 + side2 + side3) / 2
    print('Площадь треугольника: ', (half_per * (half_per - side1) * (half_per - side2) * (half_per - side3)) ** 0.5)
elif figure == 'ПРЯМОУГОЛЬНИК':
    side1 = int(input('Введите длину первой стороны: '))
    side2 = int(input('Введите длину второй стороны: '))
    print('Площадь прямоугольника: ', side1 * side2)
```
