<h3>Задание 1</h3>

```python
word = input('Введите слово: ')

if len(word) % 2 == 0:
    print(word[int(len(word) / 2) - 1] + word[int(len(word) / 2)])
else:
    print(word[int(len(word) // 2)])
```

```python
full_sum = 0
number = 1

while number != 0:
    number = int(input('Введите число: '))
    full_sum += number
    
print(full_sum)
```

<h3>Задание 3</h3>

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(girls) == len(boys):
    print('Идеальные пары:')
    sorted_girls = sorted(girls)
    sorted_boys = sorted(boys)
    
    i = 0
    
    while i < len(sorted_girls):
        print(f'{sorted_girls[i]} и {sorted_boys[i]}')
        i += 1
else:
    print('Внимание, кто-то может остаться без пары!')
```

<h3>Задание 4</h3>

```python
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

for country in countries_temperature:
    country_name = country[0]
    medium_temp = 0
    for temperature in country[1]:
        medium_temp += temperature
    medium_temp = medium_temp / len(country[1])
    medium_temp_cels = round((medium_temp - 32) * (5/9), 1)
    print(f'{country_name} - {medium_temp_cels} C')
```

<h3>Задание 5</h3>

```python
import re
number_letters = ['А', 'В', 'Е', 'К', 'М', 'Н', 'О', 'Р', 'С', 'Т', 'У', 'Х']

car_ids = ['А222ВС96', 'АБ22ВВ193']

for number in car_ids:
    if (number[0]).upper() in number_letters:
        if re.match(r"[0-9]{3}", str(number[1:4])):
            if (number[4]).upper() in number_letters and (number[5]).upper() in number_letters:
                if re.match(r"[0-9]{2,3}", number[6:]):
                    print(f'Номер {number[0:6]} валиден. Регион: {number[6:]}')
                else:
                    print(f'Номер {number} не валиден')
            else:
                print(f'Номер {number} не валиден')
        else:
            print(f'Номер {number} не валиден')
    else:
        print(f'Номер {number} не валиден')
```

<h3>Задание 6</h3>

```python
stream = [
    'user100,2022-01-01;150',
    'user99,2022-01-07;205',
    'user1001,2022-03-29;81'
]

new_data = []
unique_users = set()
full_views = 0

for row in stream:
    user = (row.split(','))[0]
    date = (((row.split(','))[1]).split(';'))[0]
    views = int((row.split(';'))[1])
    new_data.append([user, date, views])
    unique_users.add(user)
    full_views += views


avg_views = round((full_views / len(unique_users)), 2)

print(f'Среднее количество просмотров на уникального пользователя: {avg_views}')
```
