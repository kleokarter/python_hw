<h3>Задание 2</h3>

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

queries_stat = dict()

for query in queries:
    query_size = len(query.split(' '))
    if query_size in queries_stat.keys():
        queries_stat[query_size] += 1
    else:
        queries_stat[query_size] = 1

queries_all = len(queries)

for query_size in queries_stat:
    queries_stat[query_size] = round((queries_stat[query_size] / queries_all) * 100, 2)
    print(f'Поисковых запросов, содержащих {query_size} слов(а): {queries_stat[query_size]}')
```

```python
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},   
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},    
    ]
}

person_quantity = int(input('Введите количество порций: '))

new_cook_book = dict()

for item in cook_book:
    for ingr in cook_book[item]:
        new_ingr = ingr['ingridient_name'].capitalize() + "_" + ingr['measure']
        new_ingr_values = [ingr['ingridient_name'].capitalize(),
                           ingr['quantity'] * person_quantity, 
                           ingr['measure']]
        if new_ingr in new_cook_book:
            new_cook_book[new_ingr][1] += new_ingr_values[1]
        else:
            new_cook_book[new_ingr] = new_ingr_values

for item in new_cook_book:
    print(f'{new_cook_book[item][0]}: {new_cook_book[item][1]} {new_cook_book[item][2]}')
```
